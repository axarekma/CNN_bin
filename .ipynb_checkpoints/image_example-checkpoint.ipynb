```python
# Basic Template
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os
from skimage.measure import compare_psnr
import cnnbin
from cnnbin.utils import psnr
```

### Gray value image

```python
sigma = 30
image_ref = np.mean(imageio.imread('data/monarch.bmp'),2)
image = image_ref+sigma * np.random.standard_normal(image_ref.shape)

binned_ref = np.mean(cnnbin.split_diagonal(image_ref),0)
binned_mean = np.mean(cnnbin.split_diagonal(image),0)
```

```python
block_size = (256,256)
batch_size = 8
net = cnnbin.CNNbin(depth=4, start_filts=48, block_size=block_size,batch_size = batch_size )
size = net.summary()
```

```python
net.train_random(image, num_epochs=10, batch_size=6,
                 num_batches=2, learning_rate=1e-3, alpha=0.99)
net.plot_train()

image_filter_fuse = net.filter(image, sampling=1.01)
image_filter = net.filter_patch(image)

plt.figure(figsize=(13, 5))
for i, im in enumerate([binned_mean, image_filter, image_filter_fuse]):
    plt.subplot(1, 3, 1+i)
    plt.imshow(np.clip(binned_ref/255, 0, 1))
    plt.title(f'PSNR {psnr(im,binned_ref):.2f}')
```

```python
net2 = cnnbin.CNN_filter(depth=4, start_filts=48, block_size = block_size)
net2.train_split(image,num_epochs= 2,batch_size = 12, lr=1e-3,alpha = 0.99)
net2.plot_train()
net2.train_split(image,num_epochs= 2,batch_size = 12, lr=1e-4,alpha = 0.99)
net2.plot_train()

image_filter_fuse = net2.filter(image,sampling=1.01)
image_filter = net2.filter_patch(image)

plt.figure(figsize=(13,5))
for i,im in enumerate([image_bin,image_filter,image_filter_fuse]):
    plt.subplot(1,3,1+i)    
    plt.imshow(np.clip(im/255,0,1))
    plt.title(f'PSNR {psnr(im,im_ref):.2f}')
```

### Color image

```python
sigma = 30
image_ref =imageio.imread('data/monarch.bmp')
image = image_ref+sigma * np.random.standard_normal(image_ref.shape)
block_size = (256,256)

binned_ref = np.mean(cnnbin.split_diagonal_rgb(image_ref),0)
binned_mean = np.mean(cnnbin.split_diagonal_rgb(image),0)
```

```python
net = cnnbin.CNN_filter(multichannel = True, depth=4, start_filts=48, block_size = block_size)
net.train_random(image,num_epochs= 2,batch_size = 6,num_batches = 2, lr=1e-3,alpha = 0.99)
net.plot_train()
```

```python
net.train_random(image,num_epochs= 2,batch_size = 6,num_batches = 2, lr=1e-4,alpha = 0.99)
net.plot_train()

image_filter_fuse = net.filter(image,sampling=1.01)
image_filter = net.filter_patch(image)

plt.figure(figsize=(13,5))
for i,im in enumerate([binned_mean,image_filter,image_filter_fuse]):
    plt.subplot(1,3,1+i)    
    plt.imshow(np.clip(im/255,0,1))
    plt.title(f'PSNR {psnr(im,binned_ref):.2f}')
```

```python
net2 = cnnbin.CNN_filter(multichannel = True, depth=4, start_filts=48, block_size = block_size)

net2.train_split(image,num_epochs= 2,batch_size = 12, lr=1e-3,alpha = 0.99)
net2.plot_train()
```

```python
net2.train_split(image,num_epochs= 2,batch_size = 12, lr=1e-4,alpha = 0.99)
net2.plot_train()

image_filter_fuse = net2.filter(image,sampling=1.01)
image_filter = net2.filter_patch(image)

plt.figure(figsize=(13,5))
for i,im in enumerate([binned_mean,image_filter,image_filter_fuse]):
    plt.subplot(1,3,1+i)    
    plt.imshow(np.clip(im/255,0,1))
    plt.title(f'PSNR {psnr(im,binned_ref):.2f}')
```

```python
sigma = 80
images_ref =[
    cnnbin.utils.pad2bin(imageio.imread(x)) for x in ['data/barbara.bmp','data/monarch.bmp','data/comic.bmp']
]
images = [x + sigma * np.random.standard_normal(x.shape) for x in images_ref]

for x in images:
    print(x.shape)
```

```python
block_size = (256,256)
net = cnnbin.CNN_filter(multichannel = True, depth=4, start_filts=48, block_size = block_size)
```

```python
net.train_list(images,num_epochs= 50,batch_size = 8,num_batches = 4, lr=1e-3,alpha = 0.99)
net.plot_train()
```

```python
for img,ref in zip(images,images_ref):
    binned_ref = np.mean(cnnbin.split_diagonal_rgb(ref),0)
    binned_mean = np.mean(cnnbin.split_diagonal_rgb(img),0)

    image_filter_fuse = net.filter(img,sampling=1.1)
    image_filter = net.filter_patch(img)

    plt.figure(figsize=(13,5))
    for i,im in enumerate([binned_mean,image_filter,image_filter_fuse]):
        plt.subplot(1,3,1+i)    
        plt.imshow(np.clip(im/255,0,1))
        plt.title(f'PSNR {psnr(im,binned_ref):.2f}')
```
