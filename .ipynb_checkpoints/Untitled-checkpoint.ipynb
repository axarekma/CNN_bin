```python
# Basic Template
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import cnnbin
import ncxt_utils
```

```python
from ncxt_utils.mrc import read_mrc
from cnnbin.dataset import N2NBlocks
FOLDER = 'd:/2018/CNN_denoise/legros_695_6/AREC_Split/'
stack1 = read_mrc(FOLDER+'fbp_stack1.mrc')
stack2 = read_mrc(FOLDER+'fbp_stack2.mrc')
scale = np.mean(stack1)

pad = 150
pady = 200
stack1_train = stack1[pady:-pady,pad:-pad,pad:-pad]
stack2_train = stack2[pady:-pady,pad:-pad,pad:-pad]
ncxt_utils.mrc.MRC_static(stack1_train)

block_shape = (32,64,64)
dataset = N2NBlocks(stack1_train,stack2_train,block_shape = block_shape,totorch=True)
print(dataset[0][0].shape)
print(len(dataset))
```

```python
net = cnnbin.CNN_filter3D()
```

```python
net.train(stack1/scale,stack2/scale,num_epochs= 5,num_batches = 10,block_shape = block_shape)
net.plot_train()
```

```python
net.train(stack1/scale,stack2/scale,num_epochs= 10,block_shape = (48,48,48),lr = 1e-4)
net.plot_train()
```

```python

pady = 200
stack1_out = stack1[pady:-pady,:]
stack2_out = stack2[pady:-pady,:]
print(stack1_out.shape)
ncxt_utils.mrc.MRC_static(stack1_out)
```

```python
filtered = net.filter_stack(stack1/scale,stack2/scale,block_shape = (48,48,48))*scale
```

```python
import imageio
from cnnbin.utils import split_rgb
folder = 'd:/2018/CNN_denoise/set14/'
image = imageio.imread(folder + 'monarch.bmp')
noisy = image + 35 * np.random.standard_normal(image.shape)
im1,im2 = split_rgb(noisy,mode = 'diag')

net = cnnbin.CNN_filter(in_channels=3)
im_filter = net.train(im1,im2,num_epochs= 5,lr=1e-3)
net.plot_train()

plt.figure(figsize=(13,5))
for i,im in enumerate([(im1+im2)/2,im_filter]):
    plt.subplot(1,3,1+i)
    plt.imshow(np.clip(im/255,0,1))
```

```python
index = 5
im1,im2 = stack1[index],stack2[index]
im_filter = net.train(im1,im2,num_epochs= 5,lr=1e-3)
net.plot_train()
```

```python
tomolim = (np.min(im1),np.max(im1))
plt.figure(figsize=(13,5))
for i,im in enumerate([im1,im2,im_filter]):
    plt.subplot(1,3,1+i)
    plt.imshow(im[100:200,100:200],clim=tomolim)
```
